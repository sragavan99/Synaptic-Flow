```
# magnitude pruning

for path in os.listdir('ExpResults'):
  wgt_file = 'ExpResults/' + path + '/model.pt'
  for file in os.listdir('ExpResults/' + path):
    if file.startswith('post-final'):
      mask_file = 'ExpResults/' + path + '/' + file
      break
  print(path)
  #print("At init")
  #visualise(wgt_file=wgt_file, mask_file=wgt_file, arch='wide-resnet32', rescale=True, bins=100)
  #print("After pruning")
  visualise(wgt_file=wgt_file, mask_file=mask_file, arch='wide-resnet32', rescale=True, bins=100)
```

```
# more magnitude pruning

for path in os.listdir('ExpResults'):
  wgt_file = 'ExpResults/' + path + '/multishot/results/model.pt'
  print('ExpResults/' + path + '/multishot/results')
  for file in os.listdir('ExpResults/' + path + '/multishot/results'):
    if file.startswith('post-final'):
      mask_file = 'ExpResults/' + path + '/multishot/results/' + file
      break
  print(path)
  arch = 'vgg16-bn' if path.startswith('vgg') else 'wide-resnet32'
  
  visualise(wgt_file=wgt_file, mask_file=mask_file, arch=arch, rescale=True, bins=100)
```
